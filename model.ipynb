```python
import yfinance as yf
```

```python
# download the data
sp500 = yf.Ticker("^GSPC")
sp500 = sp500.history(period="max")
sp500
```

```python
sp500.index
```

```python
# plot closing price against dates (index)
sp500.plot(y="Close", use_index=True)
```

```python
# remove the dividends and stock splits columns
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
# create a new collumn for tomorrow's price from shifting close price up by 1
sp500["Tomorrow"] = sp500["Close"].shift(-1)
sp500
```

```python
# create a new column Target to indicate if tomorrow's price is greater than close
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
sp500
```

```python
# remove data before 1990
sp500 = sp500.loc["1990-01-01":].copy()
```
